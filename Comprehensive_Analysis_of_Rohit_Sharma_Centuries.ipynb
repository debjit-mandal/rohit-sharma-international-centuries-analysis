# Comprehensive Analysis of Rohit Sharma's Centuries

## Century Distribution by Match Type

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data/Rohit_Sharma_Centuries.csv'
df = pd.read_csv(file_path)

match_type_counts = df['Type of Match'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=match_type_counts.index, y=match_type_counts.values, palette='viridis')
plt.title('Century Distribution by Match Type')
plt.xlabel('Match Type')
plt.ylabel('Number of Centuries')
plt.show()
```

### Insights
The above plot shows the distribution of Rohit Sharma's centuries across different match formats. We can observe which format he has been most prolific in terms of scoring centuries.

## Centuries Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/Rohit_Sharma_Centuries.csv'
df = pd.read_csv(file_path)

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

df['Year'] = df['Date'].dt.year

centuries_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(centuries_per_year.index, centuries_per_year.values, marker='o', linestyle='-', color='b')
plt.title("Rohit Sharma's Centuries Over Time")
plt.xlabel('Year')
plt.ylabel('Number of Centuries')
plt.grid(True)
plt.show()
```

### Insights
The plot above illustrates the trend of centuries scored by Rohit Sharma over the years. It highlights the periods where he was particularly successful and any fluctuations in his century-scoring frequency.

## Venue Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/Rohit_Sharma_Centuries.csv'
df = pd.read_csv(file_path)

venue_counts = df['Venue'].value_counts()

plt.figure(figsize=(12, 8))
venue_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Venues Where Rohit Sharma Scored the Most Centuries')
plt.xlabel('Venue')
plt.ylabel('Number of Centuries')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insights
The bar plot above highlights the top 10 venues where Rohit Sharma has scored the most centuries. This analysis helps in understanding his favorite grounds and his performance trends at different locations.
